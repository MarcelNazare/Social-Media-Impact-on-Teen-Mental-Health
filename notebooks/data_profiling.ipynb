## Checking for Outliers in our Data


```python
def detect_outliers_iqr(data: pl.DataFrame) -> pl.DataFrame:
    numeric_cols = [col for col, dtype in zip(data.columns, data.dtypes) 
                    if dtype in (pl.Int64, pl.Int32, pl.Float64, pl.Float32)]
    
    outlier_flags = []
    
    for col in numeric_cols:
        q1 = data.select(pl.col(col).quantile(0.25)).item()
        q3 = data.select(pl.col(col).quantile(0.75)).item()
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        # Boolean mask for outliers in this column
        mask = (pl.col(col) < lower_bound) | (pl.col(col) > upper_bound)
        outlier_flags.append(mask.alias(f"{col}_is_outlier"))
    
    # Combine original data with outlier flags
    outlier_df = data.with_columns(outlier_flags)
    
    # Count and print the number of outliers for each numeric column
    for col in numeric_cols:
        outlier_col = f"{col}_is_outlier"
        count = outlier_df.select(pl.col(outlier_col).sum()).item()
        print(f"Number of outliers in {col}: {count}")
    
    return outlier_df
```

```python
detect_outliers_iqr(df)
```
