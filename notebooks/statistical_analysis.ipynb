### Descriptive Statistics 

#### Central tendency and dispersion
- **Mean Social Media Usage** - The average time spent online across the entire dataset.
- **Standard Deviation of Stress** - The degree of variation in stress levels among respondents.

```python
def descriptive_analytics(data):
    # Calculate measures of central tendency
    mean = ps.arithmetic_mean(data)
    median = ps.median(data)
    mode = ps.mode(data)
    # Calculate measures of dispersion
    variance = ps.variance(data,population=False)
    std_dev = ps.standard_deviation(data, population=False)
    range = ps.midrange(data)
    context = {
        "mean": mean,
        "median": median,
        "mode": mode,
        "variance": variance,
        "std_dev": std_dev,
        "range": range  
    }

    return context  
```

```python
# Run descriptive analytics for all numerical columns and display nicely
results = {}
numeric_cols = [col for col, dtype in zip(df.columns, df.dtypes) if dtype in (pl.Int64, pl.Int32, pl.Float64, pl.Float32)]
for col in numeric_cols:
    stats = descriptive_analytics(df[col].to_list())
    results[col] = stats
```

```python
# Prepare data for DataFrame
stats_rows = []
for col, stats in results.items():
    row = {"column": col}
    row.update(stats)
    stats_rows.append(row)

stats_df = pl.DataFrame(stats_rows)
stats_df
```

# Correlation Analysis

#### Correlation Matrix 
- **Screen Time Before Sleep vs. Sleep Hours** - Analyzes the relationship between late-night device use and rest hours.
- **Social Media Hours vs. Anxiety** - Examines the link between time spent on platforms and anxiety scores.

```python
# Checking if there is any correlation between screen time before sleep and sleep hours 
screen_time = df['screen_time_before_sleep'].to_list()
sleep_hours = df['sleep_hours'].to_list()


# pearson correlation package imported from py_stats as ps.pearson_correlation
correlation = ps.pearson_correlation(screen_time, sleep_hours)
# 0.0<0.1 = No correlation 0.1<0.3 = little correlation 03<05 medium correlation 05<07 high correlation 0.7 < 1 very high correlation
if correlation == 0.0 or correlation <= 0.1:
    print("No Correlation")
elif correlation == 0.1 or correlation <= 0.3:
    print("Little Correlation")
elif correlation == 0.3 or correlation <= 0.5:
    print("Medium Correlation")
elif correlation == 0.5 or correlation <= 0.7:
    print("High Correlation")
else:
    print("Very High Correlation")
print(f"Pearson Correlation between Screen Time Before Sleep and Sleep Hours: {correlation}")   
```

```python
# Checking if there is any correlation between social media hours and anxiety levels    
daily_social_media_hours = df['daily_social_media_hours'].to_list()
anxiety_level = df['anxiety_level'].to_list()

# pearson correlation package imported from py_stats as ps.pearson_correlation
correlation = ps.pearson_correlation(daily_social_media_hours, anxiety_level)

# 0.0<0.1 = No correlation 0.1<0.3 = little correlation 03<05 medium correlation 05<07 high correlation 0.7 < 1 very high correlation
if correlation == 0.0 or correlation <= 0.1:
    print("No Correlation")
elif correlation == 0.1 or correlation <= 0.3:
    print("Little Correlation")
elif correlation == 0.3 or correlation <= 0.5:
    print("Medium Correlation")
elif correlation == 0.5 or correlation <= 0.7:
    print("High Correlation")
else:
    print("Very High Correlation")
print(f"Pearson Correlation between daily socialy media hours and anxiety level: {correlation}")   
```
