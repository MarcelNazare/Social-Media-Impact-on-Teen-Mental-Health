### Descriptive Statistics 

#### Central tendency and dispersion
- **Mean Social Media Usage** - The average time spent online across the entire dataset.
- **Standard Deviation of Stress** - The degree of variation in stress levels among respondents.

```python
def descriptive_analytics(data):
    # Calculate measures of central tendency
    mean = ps.arithmetic_mean(data)
    median = ps.median(data)
    mode = ps.mode(data)
    # Calculate measures of dispersion
    variance = ps.variance(data,population=False)
    std_dev = ps.standard_deviation(data, population=False)
    range = ps.midrange(data)
    context = {
        "mean": mean,
        "median": median,
        "mode": mode,
        "variance": variance,
        "std_dev": std_dev,
        "range": range  
    }

    return context  
```

```python
# Run descriptive analytics for all numerical columns and display nicely
results = {}
numeric_cols = [col for col, dtype in zip(df.columns, df.dtypes) if dtype in (pl.Int64, pl.Int32, pl.Float64, pl.Float32)]
for col in numeric_cols:
    stats = descriptive_analytics(df[col].to_list())
    results[col] = stats
```

```python
# Prepare data for DataFrame
stats_rows = []
for col, stats in results.items():
    row = {"column": col}
    row.update(stats)
    stats_rows.append(row)

stats_df = pl.DataFrame(stats_rows)
stats_df
```

```python
import plotly.graph_objects as go

def save_df_as_table(df, filename='table', title=''):
    fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(df.columns),
            fill_color='#2c3e50',
            font=dict(color='white', size=13),
            align='center',
            height=40
        ),
        cells=dict(
            values=[df[col] for col in df.columns],
            fill_color='#f9f9f9',
            font=dict(color='#2c3e50', size=12),
            align='center',
            height=35
        )
    )])

    fig.update_layout(
        title=title,
        margin=dict(l=20, r=20, t=60, b=20)
    )

    #fig.write_image(f'{filename}.png')   # save as PNG
    fig.write_html(f'../visualizations/{filename}.html') # save as interactive HTML
    fig.show()
```

# Correlation Analysis

#### Correlation Matrix 
- **Screen Time Before Sleep vs. Sleep Hours** - Analyzes the relationship between late-night device use and rest hours.
- **Social Media Hours vs. Anxiety** - Examines the link between time spent on platforms and anxiety scores.

```python
# Checking if there is any correlation between screen time before sleep and sleep hours 
screen_time = df['screen_time_before_sleep'].to_list()
sleep_hours = df['sleep_hours'].to_list()


# pearson correlation package imported from py_stats as ps.pearson_correlation
correlation = ps.pearson_correlation(screen_time, sleep_hours)
# 0.0<0.1 = No correlation 0.1<0.3 = little correlation 03<05 medium correlation 05<07 high correlation 0.7 < 1 very high correlation
if correlation == 0.0 or correlation <= 0.1:
    print("No Correlation")
elif correlation == 0.1 or correlation <= 0.3:
    print("Little Correlation")
elif correlation == 0.3 or correlation <= 0.5:
    print("Medium Correlation")
elif correlation == 0.5 or correlation <= 0.7:
    print("High Correlation")
else:
    print("Very High Correlation")
print(f"Pearson Correlation between Screen Time Before Sleep and Sleep Hours: {correlation}")   
```

```python
# Checking if there is any correlation between social media hours and anxiety levels    
daily_social_media_hours = df['daily_social_media_hours'].to_list()
anxiety_level = df['anxiety_level'].to_list()

# pearson correlation package imported from py_stats as ps.pearson_correlation
correlation = ps.pearson_correlation(daily_social_media_hours, anxiety_level)

# 0.0<0.1 = No correlation 0.1<0.3 = little correlation 03<05 medium correlation 05<07 high correlation 0.7 < 1 very high correlation
if correlation == 0.0 or correlation <= 0.1:
    print("No Correlation")
elif correlation == 0.1 or correlation <= 0.3:
    print("Little Correlation")
elif correlation == 0.3 or correlation <= 0.5:
    print("Medium Correlation")
elif correlation == 0.5 or correlation <= 0.7:
    print("High Correlation")
else:
    print("Very High Correlation")
print(f"Pearson Correlation between daily socialy media hours and anxiety level: {correlation}")   
```

### RFM Analysis 
- **User Engagement Segmentation** - Categorizing users by usage frequency (Daily Hours) and intensity (Addiction Level).

```python
import pandas as pd
df = df.to_pandas()

# Score Addiction Level (higher = worse, so rank descending)
df['A_score'] = pd.qcut(df['addiction_level'], q=5, labels=[1, 2, 3, 4, 5])

# Score Daily Hours (higher = more engaged)
df['F_score'] = pd.qcut(df['daily_social_media_hours'], q=5, labels=[1, 2, 3, 4, 5])

# Score Academic Performance (higher = more impacted negatively)
df['M_score'] = pd.qcut(df['academic_performance'], q=5, labels=[5, 4, 3, 2, 1])  # reversed
```

```python
df['A_score'] = df['A_score'].astype(int)
df['F_score'] = df['F_score'].astype(int)
df['M_score'] = df['M_score'].astype(int)

df['AFM_score'] = df['A_score'].astype(str) + df['F_score'].astype(str) + df['M_score'].astype(str)
df['AFM_total'] = df['A_score'] + df['F_score'] + df['M_score']
```

```python
def segment_user(score):
    if score >= 12:
        return 'High Risk'
    elif score >= 8:
        return 'Moderate Risk'
    elif score >= 5:
        return 'Low Engagement'
    else:
        return 'Healthy User'

df['segment'] = df['AFM_total'].apply(segment_user)
```

```python
segment_summary = df.groupby('segment').agg(
    count=('segment', 'count'),
    avg_daily_social_media_usage_hours=('daily_social_media_hours', 'mean'),
    avg_addiction_level=('addiction_level', 'mean'),
    avg_academic_perfomance=('academic_performance', 'mean'),
    avg_sleep_hours=('sleep_hours', 'mean')
).round(2).reset_index()

segment_summary
```

```python
import plotly.graph_objects as go

fig_table = go.Figure(data=[go.Table(
    header=dict(
        values=[
            'Segment', 'Count', 'Avg Daily Social Media Usage Hours',
            'Avg Addiction Level', 'Avg Academic Perfomance', 'Avg Sleep Hours'
        ],
        fill_color='#2c3e50',
        font=dict(color='white', size=13),
        align='center',
        height=40
    ),
    cells=dict(
        values=[
            segment_summary['segment'],
            segment_summary['count'],
            segment_summary['avg_daily_social_media_usage_hours'],
            segment_summary['avg_addiction_level'],
            segment_summary['avg_academic_perfomance'],
            segment_summary['avg_sleep_hours']
        ],
        fill_color=[
            ['#e74c3c' if s == 'High Risk'
             else '#e67e22' if s == 'Moderate Risk'
             else '#f1c40f' if s == 'Low Engagement'
             else '#2ecc71' for s in segment_summary['segment']]
        ],
        font=dict(color='white', size=12),
        align='center',
        height=35
    )
)])

fig_table.update_layout(
    title='Segmentation Summary',
    title_font=dict(size=16),
    margin=dict(l=20, r=20, t=60, b=20)
)

#fig_table.write_image('../visualization/Segment summary table.png')  # save as PNG
# OR
fig_table.write_html('../visualizations/Segment summary table.html')  # save as interactive HTML

fig_table.show()
```
