```python
def drop_nulls(df, subset=None):
    """
    Drop rows with null values from the DataFrame.
    
    Parameters:
    df (pl.DataFrame): The DataFrame to clean.
    subset (list of str, optional): Columns to consider for null values. If None, considers all columns.
    
    Returns:
    pl.DataFrame: DataFrame with null rows dropped.
    """
    return df.drop_nulls(subset=subset)
```

```python
def clean_and_report_changes(df):
    """
    Drops null values and duplicates from the DataFrame, and reports row changes at each step.
    Returns the cleaned DataFrame and a dict with row counts.
    """
    initial_rows = df.shape[0]
    df_no_nulls = df.drop_nulls()
    after_nulls = df_no_nulls.shape[0]
    df_no_dupes = df_no_nulls.unique()
    after_dupes = df_no_dupes.shape[0]
    changes = {
        'initial_rows': initial_rows,
        'after_drop_nulls': after_nulls,
        'after_drop_duplicates': after_dupes,
        'nulls_dropped': initial_rows - after_nulls,
        'duplicates_dropped': after_nulls - after_dupes
    }
    return df_no_dupes, changes

# Example usage:
# cleaned_df, row_changes = clean_and_report_changes(df)
# print(row_changes)
```

## Removing Outliers

```python
import polars as pl

def remove_outliers_iqr(data: pl.DataFrame) -> pl.DataFrame:
    """
    Removes rows containing outliers in any numeric column using the IQR method.
    Returns a DataFrame with outliers removed.
    """
    numeric_cols = [col for col, dtype in zip(data.columns, data.dtypes)
                    if dtype in (pl.Int64, pl.Int32, pl.Float64, pl.Float32)]
    mask = None
    for col in numeric_cols:
        q1 = data.select(pl.col(col).quantile(0.25)).item()
        q3 = data.select(pl.col(col).quantile(0.75)).item()
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        col_mask = (data[col] >= lower_bound) & (data[col] <= upper_bound)
        mask = col_mask if mask is None else (mask & col_mask)
    cleaned = data.filter(mask) if mask is not None else data
    return cleaned

# Example usage:
# df = remove_outliers_iqr(df)
```
