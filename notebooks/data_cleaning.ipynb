# Data Cleaning

```python
# lets import our data
from load_data import load_unprocessed_data
df = load_unprocessed_data()
```

```python
def drop_nulls(df, subset=None):
    """
    Drop rows with null values from the DataFrame.
    
    Parameters:
    df (pl.DataFrame): The DataFrame to clean.
    subset (list of str, optional): Columns to consider for null values. If None, considers all columns.
    
    Returns:
    pl.DataFrame: DataFrame with null rows dropped.
    """
    return df.drop_nulls(subset=subset)
```
